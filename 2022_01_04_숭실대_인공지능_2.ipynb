```python
import requests # 엔터키(페이지 요청)
from bs4 import BeautifulSoup # 파이썬에서 HTML을 다루는 기능
import time
from tqdm import tqdm # for문의 진행상황 확인 가능
```

```python
total = []

for n in tqdm(range(1,3992,10)):

    url = requests.get("https://search.naver.com/search.naver?where=news&sm=tab_pge&query=%EC%86%90%ED%9D%A5%EB%AF%BC&sort=0&photo=0&field=0&pd=0&ds=&de=&cluster_rank=147&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:all,a:all&start={}".format(n))
    html = BeautifulSoup(url.text)
    contents = html.find('ul', class_ = 'list_news').find_all('li', class_='bx')

    for i in contents:
        title = i.find('a', class_='news_tit').text
        summary = i.find('div', class_='dsc_wrap').text
        link = i.find('a', class_ = 'news_tit')['href']
        total.append([title, summary, link])

    time.sleep(0.5)
```

```python
len(total)
```

```python
df = pd.DataFrame(total, columns = ['제목','내용','링크'])
df.to_excel('son.xlsx')
df
```

```python
import requests # 엔터키(페이지 요청)
from bs4 import BeautifulSoup # 파이썬에서 HTML을 다루는 기능
import time
from tqdm import tqdm # for문의 진행상황 확인 가능
```

```python
# 코스피 크롤링 코드

import pandas as pd

url = requests.get("https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page=1")
html = BeautifulSoup(url.text)

total = []

def create_table(html, name):
    table = html.find('table', class_ = 'type_2')
    table = pd.read_html(str(table))[0]

    table = table[table['종목명'].notnull()]

    del table['N']
    del table['토론실']

    table['종류'] = ['{}'.format(name)]*len(table)

    return table
```

```python
url = requests.get("https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page=1")
html = BeautifulSoup(url.text)
kospi_last_page = int(html.find('td', class_ = 'pgRR').find('a')['href'].split('=')[-1])

url = requests.get("https://finance.naver.com/sise/sise_market_sum.naver?sosok=1&page=1")
html = BeautifulSoup(url.text)
kosdaq_last_page = int(html.find('td', class_ = 'pgRR').find('a')['href'].split('=')[-1])

total = []

for n in tqdm(range(1, kospi_last_page + 1)):
  url = requests.get("https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page={}".format(n))
  html = BeautifulSoup(url.text)
  total.append(create_table(html, 'KOSPI'))
  time.sleep(0.3)

for n in tqdm(range(1, kosdaq_last_page + 1)):
  url = requests.get("https://finance.naver.com/sise/sise_market_sum.naver?sosok=1&page={}".format(n))
  html = BeautifulSoup(url.text)
  total.append(create_table(html, 'KOSDAQ'))
  time.sleep(0.3)

stock = pd.concat(total, ignore_index=True)
stock.to_excel('stock.xlsx')
stock
```

```python
# 정형 데이터 : 엑셀(머신러닝 모델)
# 비정형 데이터 : 이미지, 음성, 텍스트(딥러닝 모델)

import numpy as np
import matplotlib.pyplot as plt

x = np.array([2,4,6,8])
y = np.array([30,75,55,95])

# 사람: 6, 35

a = 0
b = 0
lr = 0.005

for i in range(1, 3001):

    y_hat = a*x + b
    error = y_hat - y

    a_diff = sum(2*error*x)
    b_diff = sum(2*error*1)

    a = a - a_diff*lr
    b = b - b_diff*lr

    print('{}학습 // 기울기 {} // 절편 {}'.format(i,a,b))
```

```python
plt.scatter(x,y)
plt.plot(x, a*x+b, 'r')
```
