```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([2, 4, 6, 8]) # 공부시간
x2 = np.array([1, 2, 1, 3]) # 문제집 푼 개수
y = np.array([30, 75, 55, 95]) # 성적

# 사람 : 6, 35

a1 = 0
a2 = 0
b = 0
lr = 0.005

for i in range(1, 3001):

  y_hat = a1 * x1 + a2 * x2 + b
  error = y_hat - y
  
  a1_diff = sum(2 * error * x1)
  a2_diff = sum(2 * error * x2)
  b_diff = sum(2 * error * 1)

  a1 = a1 - a1_diff * lr
  a2 = a2 - a2_diff * lr
  b = b - b_diff * lr

  print('{}학습 // 기울기1 {} // 기울기2 {} // 절편 {}'.format(i, a1, a2, b))
```

```python
plt.scatter(x1, y)
plt.plot(x1, a1*x1+a2*x2+b, 'r')
```

```python
abs_diff = abs(fruits - banana_mean)
abs_mean = np.mean(abs_diff, axis = (1,2))

# abs_mean # 가장 작은 값 1등부터 100까지는 사과사진일것같다.

apple_index = np.argsort(abs_mean)[:100]

fig, axs = plt.subplots(10, 10, figsize = (10, 10))

for i in range(10):
  for j in range(10):
    axs[i, j].imshow(fruits[apple_index[i * 10 + j]], cmap = 'gray_r')
    axs[i, j].axis('off')

plt.show()
```

```python
# 라벨링을 하고싶은지 몇가지 카테고리로 할지 모를때

box = []

for i in range(2, 7):
  
  model = KMeans(n_clusters=i)
  model.fit(fruits2)

  box.append(model.inertia_)

plt.plot(range(2,7), box)
plt.show()
```
